```python
from pycaret import classification as clf
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
```

```python
census_data = pd.read_csv('data/census_no_spaces.csv')
```

```python
census_data.columns
```

```python
census_data.salary.unique()
```

```python
df_train, df_holdout = train_test_split(census_data, train_size=0.8, stratify=census_data.salary)
```

```python
# unbanceled dataset
```

```python
cat_features = [
    "workclass",
    "education",
    "marital-status",
    "occupation",
    "relationship",
    "race",
    "sex",
    "native-country",
]

clf.setup(data=df_train, 
          target='salary', 
          categorical_features=cat_features, 
          train_size=0.8, remove_multicollinearity=True, fix_imbalance=True, 
         )
```

```python
best_model = clf.compare_models(fold=10, )
```

```python
clf.evaluate_model(best_model)
```

```python
df_holdout.columns
```

```python
finalized_model = clf.finalize_model(best_model)
```

```python
clf.evaluate_model(finalized_model)
```

```python
y_pred = clf.predict_model(finalized_model, data=df_holdout.drop(columns=['salary']))['Label']
```

```python
y = df_holdout.salary
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y, y_pred))
```
